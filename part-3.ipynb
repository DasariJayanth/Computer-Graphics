# Using Sobel

```python
%matplotlib inline
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/cg_image_3.webp')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)

#sobel
sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=3)
sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=3)

sobel = sobelx + sobely

print("\033[1m" + "Original Image: ")
cv2_imshow(img)
print("\033[1m" + "Sobel X: ")
cv2_imshow(sobelx)
print("\033[1m" + "Sobel y: ")
cv2_imshow(sobely)
print("\033[1m" + "Sobel: ")
cv2_imshow(sobel)
cv2.waitKey(0)
```

# Using Prewitt

```python
import cv2
import numpy as np

img = cv2.imread('/content/cg_image_3.webp')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0) # Used to remove noise, blur the image etc. 

kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])    

img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
prewitt=img_prewittx + img_prewitty

print("\033[1m" + "Original Image: ")
cv2_imshow(img)
print("\033[1m" + "Prewitt X: ")
cv2_imshow(img_prewittx)
print("\033[1m" + "Prewitt Y: ")
cv2_imshow(img_prewitty)
print("\033[1m" + "Prewitt")
cv2_imshow(prewitt)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Using Robert

```python
import cv2
import numpy as np

img = cv2.imread('/content/cg_image_3.webp')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blured_img = cv2.GaussianBlur(gray,(3,3),0)

kernelx = np.array([[1, 0], [0, -1]])
kernely = np.array([[0, 1], [-1, 0]])
img_robertx = cv2.filter2D(blured_img, -1, kernelx)
img_roberty = cv2.filter2D(blured_img, -1, kernely)
robert=img_robertx+img_roberty

print("\033[1m" + "Original Image: ")
cv2_imshow(img)
print("\033[1m" + "Robert X: ")
cv2_imshow(img_robertx)
print("\033[1m" + "Robert Y: ")
cv2_imshow(img_roberty)
print("\033[1m" + "Robert Image: ")
cv2_imshow(robert)

cv2.waitKey(0)
```
